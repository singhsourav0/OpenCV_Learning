```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
import random

# Read the image
image = cv2.imread("ss.jpeg" , 0)

# Display the image with a title
cv2_imshow(image)

```

# Resizing the Image

```python
#Rotate and Resize
img1 = cv2.imread("ss.jpeg", -1)
img1 = cv2.resize(img1, (200, 200))
cv2_imshow(img1)
```

```python
#Custom Resizing
img2 = cv2.resize(img1, (0, 0), fx = 1.6 ,fy = .8)
cv2_imshow(img2)
```

**Rotating the Image**

```python
#Rotating
img3 = cv2.rotate(img2,cv2.ROTATE_90_CLOCKWISE)
cv2.imwrite('newss.jpeg', img3)
cv2_imshow(img3)
```
